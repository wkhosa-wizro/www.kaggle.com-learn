# Lesson 2 of Data Visualisation from Kaggle


```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import os
```

```python
# Loading data
sportify_filepath = '../input/sportify'

if not os.path.exists(sportify_filepath):
    os.symlink("../input/data-for-datavis/spotify.csv", sportify_filepath)

sportify_data = pd.read_csv(sportify_filepath, index_col="Date", parse_dates=True)

```

```python
# Ploting Data
sns.lineplot(data=sportify_data)
```

```python
# 
plt.figure(figsize=(14,6))
plt.title("Daily Global Streams of Popular Songs in 2017-2018")
sns.lineplot(data=sportify_data)
```

```python
# Each line is a column in a data table and Index is the x-axis
list(sportify_data.columns)
```

```python
plt.figure(figsize=(14,6))
plt.title("Daily Global Streams of Popular Songs in 2017-2018")
sns.lineplot(data=sportify_data['Shape of You'], label='Shape of You')
sns.lineplot(data=sportify_data['Despacito'], label='Despacito')
plt.xlabel("Date")
```
